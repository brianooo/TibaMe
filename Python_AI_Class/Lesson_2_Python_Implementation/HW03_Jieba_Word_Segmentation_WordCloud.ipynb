# Read Speech.txt file, use jieba to do word segmentation and do word frequency on each word

```python
import jieba
import json


f = open("./speech.txt", "r")
ret = f.read().strip()

## write your code here
words = jieba.cut(ret, cut_all=False)
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print(frequency)
f.close()

with open("count1.csv","w", encoding='utf-8-sig') as fd:
    ## write your code here
    fd.write('word, count\n')
    for key in frequency:
        fd.write('{}, {}\n'.format(key, frequency[key]))
    
```

# Use wordcloud to show the word frequency in the file "speech.txt"

```python
%matplotlib inline
from wordcloud import WordCloud
import matplotlib.pyplot as plt 


## write your code here

f = open('./speech.txt', 'r')

comment_words = f.read().strip() 

#print(comment_words)  

words = jieba.cut(comment_words, cut_all=False)
allwords = ''
for word in words:
    allwords += ' '+word

## show chinese in wordcloud: use WordCloud(font_path='simhei.ttf', ......)
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white',
                min_font_size = 10, font_path='simhei.ttf').generate(allwords) 

            
f.close()

# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()

```
